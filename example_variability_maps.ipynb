```python
from boela import problems
from boela.ela import vm_features, vm_triples
from boela.sample import DOE, SampleCached


problem: problems.ProblemBase = problems.f1.rosenbrock.Problem(2)
sample = SampleCached(problem, size=100, doe=DOE.RAND, options={"seed": 0})
x, f = sample.xf

triples = vm_triples.collect_space_filling(x, 2 * x.size, symmetric=True)
vm = vm_features.VariabilityMap(x, f, triples)

problem.plot(size=(5, 5), dpi=300, contour=False)
vm_triples.plot(x, triples, size=(5.2, 5), dpi=300)
vm.plot(size=(5.2, 5), dpi=300)


problem: problems.ProblemBase = problems.f1.rastrigin.Problem(2)
sample = SampleCached(problem, size=100, doe=DOE.RAND, options={"seed": 0})
x, f = sample.xf

triples = vm_triples.collect_space_filling(x, 2 * x.size, symmetric=True)
vm = vm_features.VariabilityMap(x, f, triples)

problem.plot(size=(5, 5), dpi=300, contour=False)
vm_triples.plot(x, triples, size=(5.2, 5), dpi=300)
vm.plot(size=(5.2, 5), dpi=300)
```
